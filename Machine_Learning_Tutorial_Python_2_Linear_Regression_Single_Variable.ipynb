```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
#https://github.com/codebasics/py/blob/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv
from google.colab import files
uploaded = files.upload()
df=pd.read_csv('a.csv')
df
```

```python
df.rename(columns={"per capita income (US$)": "gdp"}, inplace=True)
df
```

```python
plt.xlabel('year')
plt.ylabel('gdp')
plt.scatter(df.year,df.GDP)
```

```python
newdf=df.drop('GDP',axis='columns')
newdf
```

```python
GDP=df.GDP
GDP
```

```python
reg=linear_model.LinearRegression()
reg.fit(newdf,GDP)
reg.predict([[2020]])
```
